```python
import h5py
import os

file_lis = []
skipped_files = []
root = '/ptmp/yacheng/outthere_ssl/'
for folder_path in [f'{root}/images/north', f'{root}/images/south']:
    files = sorted([f for f in os.listdir(folder_path) if f.endswith('.h5')])
    for file in files:
        path = os.path.join(folder_path, file)
        # Skip clearly invalid files and handle open errors gracefully
        try:
            if not h5py.is_hdf5(path):
                print(f"Skipping non-HDF5 or corrupted file: {path}")
                skipped_files.append(path)
                continue
            fobj = h5py.File(path, 'r')
            file_lis.append(fobj)
        except (OSError, IOError) as e:
            # If a file is truncated or otherwise unreadable, skip it instead of crashing
            print(f"Failed to open {path}: {e}. Skipping.")
            skipped_files.append(path)
            continue

print(f"Successfully opened {len(file_lis)} files.")
print(f"Skipped {len(skipped_files)} files: {skipped_files}")
#check total num of images in all files

total_images = sum(fobj['images'].shape[0] for fobj in file_lis)
print(f"Total number of images across all opened files: {total_images}")
```
