```python
import numpy as np
import matplotlib.pyplot as plt
#use percentile normalization for better contrast
import matplotlib.colors as colors
batches =  np.load('/ptmp/yacheng/outthere_ssl/batch.npy')
embeddings = np.load('/ptmp/yacheng/outthere_ssl/embeddings.npy')
print(batches.shape,embeddings.shape)
#plot a few images from batches
#lets plot 8by8 grid of images
#only show the channel thats trained on
fig, axs = plt.subplots(8, 8, figsize=(12, 12))
for i in range(8):
    for j in range(8):
        axs[i, j].imshow(batches[i * 8 + j][2], norm=colors.Normalize(vmin=np.percentile(batches[i * 8 + j][2], 1), vmax=np.percentile(batches[i * 8 + j][2], 99)))  # use percent
        axs[i, j].axis('off')
plt.show()
```

```python
#try to get PCA and UMAP embeddings
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
embeddings_2d = pca.fit_transform(embeddings)
print(embeddings_2d.shape)

```
